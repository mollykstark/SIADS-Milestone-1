# MET & Chicago Institute of Art Collections - MatPlotLib Exploratory Visualizations

## Country Distribution Charts 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Load in clean MET data and count the number of times each country shows up
clean_met_df = pd.read_csv('../data/processed/met_data.csv')
clean_met_df = clean_met_df.dropna(subset=['country'])

met_country_dist = pd.DataFrame(clean_met_df['country'].value_counts())
met_country_dist = met_country_dist.reset_index()
met_country_dist = met_country_dist.head(50)
# met_country_dist = met_country_dist[met_country_dist['count'] > 25]
met_country_dist.head()
```

```python
clean_aic_df = pd.read_csv('../data/processed/aic_data.csv')

aic_country_dist = pd.DataFrame(clean_aic_df['country'].value_counts())
aic_country_dist = aic_country_dist.reset_index()
aic_country_dist = aic_country_dist.head(50)
# met_country_dist = met_country_dist[met_country_dist['count'] > 25]
aic_country_dist.head()
```

```python
met_countries = met_country_dist['country']
met_counts = met_country_dist['count'].astype(int)

opacity=0.7

#MET Country Distribution Bar Chart

fig = plt.figure(figsize=(14,8))          
plt.bar(met_countries, met_counts, color='#800000', alpha=opacity)
plt.yscale('log')
plt.xlabel('Country')
plt.ylabel('Log Count')
plt.title('Country Representation in the MET Art Collection')
plt.xticks(rotation=45, ha='right')  
# plt.yticks(np.arange(0, max(met_counts), 1000))
plt.grid(axis='y')
fig.tight_layout()
plt.show()
```

```python
aic_countries = aic_country_dist['country']
aic_counts = aic_country_dist['count'].astype(int)

opacity=0.7

#AIC Country Distribution Bar Chart

fig = plt.figure(figsize=(14,8))          
plt.bar(aic_countries, aic_counts, color='#800000', alpha=opacity)
plt.yscale('log')
plt.xlabel('Country')
plt.ylabel('Log Count')
plt.title('Country Representation in the Art Institute of Chicago Art Collection')
plt.xticks(rotation=45, ha='right')  
# plt.yticks(np.arange(0, max(aic_counts), 1000))
plt.grid(axis='y')
fig.tight_layout()
plt.show()
```
